```python
#Подлючаем библиотеки для анализа
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline 
sns.set(style="ticks")
```

```python
#Подлючаем данные
data = pd.read_csv('wine.data', sep=",")
```

```python
# Разменость датасета
total_count = data.shape
print('Всего строк: {}'.format(total_count[0]))
print('Всего колонок: {}'.format(total_count[1]))
```

```python
# Список колонок с типами данных
data.dtypes
```

```python
#Выведем первые 5 строк
data.head()
```

```python
# Проверим наличие пустых значений
for col in data.columns:
    temp_null_count = data[data[col].isnull()].shape[0]
    print('{} - {}'.format(col, temp_null_count))
```

```python
# Основные статистические характеристки набора данных
data.describe()
```

```python
#Проведем корреляционный анализ
data.corr()
```

```python
#Диаграмма рассеивания
fig, ax = plt.subplots(figsize=(10,10)) 
sns.scatterplot(ax=ax, x='OD280/OD315 of diluted wines', y='Flavanoids', data=data)
```

```python
#Диаграмма рассеивания + гистограмма
sns.jointplot(x='OD280/OD315 of diluted wines', y='Flavanoids', data=data, kind="hex")
```

```python
#"Парные диаграммы"
sns.pairplot(data)
```

```python
# Распределение параметра 'Flavanoids' сгруппированные по 'OD280/OD315 of diluted wines'.
sns.violinplot(x='OD280/OD315 of diluted wines', y='Flavanoids', data=data)
```

```python
# Треугольный вариант матрицы корреляции
mask = np.zeros_like(data.corr(), dtype=np.bool)
mask[np.triu_indices_from(mask)] = True
f, ax = plt.subplots(figsize=(15, 10))
sns.heatmap(data.corr(), mask=mask,cmap='YlGnBu', annot=True, fmt='.3f',)
```
